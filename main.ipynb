```python
#We will just keep the last 20 years and we will drop columns we won't be using

todrop = ['Y1961',
       'Y1962', 'Y1963', 'Y1964', 'Y1965', 'Y1966', 'Y1967', 'Y1968', 'Y1969',
       'Y1970', 'Y1971', 'Y1972', 'Y1973', 'Y1974', 'Y1975', 'Y1976', 'Y1977',
       'Y1978', 'Y1979', 'Y1980', 'Y1981', 'Y1982', 'Y1983', 'Y1984', 'Y1985',
       'Y1986', 'Y1987', 'Y1988', 'Y1989', 'Y1990', 'Y1991', 'Y1992', 'Area Code','Item Code', 'Element Code']

data = data.drop(columns=todrop)

#We will fill the NaN values with 0

data = data.fillna(0)
```

```python
#Also, we want to convert the column "Unit" to a float

data['Unit'] = data['Unit'].str.replace("tonnes", "").str.replace(" ", "")

data = data.rename(columns={"Unit": "NÂº of tonnes"})
data['NÂº of tonnes'] = data['NÂº of tonnes'].astype(float)
```
